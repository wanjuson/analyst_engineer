```python
# 1.데이터 불러오기
import pandas as pd
x1_train = pd.read_csv('X_train.csv', encoding = 'cp949')
x1_test = pd.read_csv('X_test.csv', encoding = 'cp949')
y1_train = pd.read_csv('y_train.csv', encoding = 'cp949')
```

```python
# 환불금액 na제거 - train
x1_train_med = x1_train['환불금액'].median()
x1_train['환불금액'][x1_train['환불금액'].isna()] = x1_train_med
# 환불금액 na제거 - test
x1_test['환불금액'][x1_test['환불금액'].isna()] = x1_train_med
```

```python
# 범주형변수 인코딩
from sklearn.preprocessing import LabelEncoder
label = LabelEncoder()
taget1 = ['주구매상품', '주구매지점']
# train
for i in taget1:
    x1_train[i] = label.fit_transform(x1_train[i])
# x1_train[taget1]
```

```python
# test
for i in taget1:
    x1_test[i] = label.fit_transform(x1_test[i])
x1_test.tail()
```
