```python
# [Python pandas] 여러개의 동일한 형태 DataFrame 합치기 : pd.concat()

pd.concat(objs,  # Series, DataFrame, Panel object
             axis=0,  # 0: 위+아래로 합치기, 1: 왼쪽+오른쪽으로 합치기
             join='outer', # 'outer': 합집합(union), 'inner': 교집합(intersection)
             join_axes=None, # axis=1 일 경우 특정 DataFrame의 index를 그대로 이용하려면 입력 (deprecated, 더이상 지원하지 않음)
             ignore_index=False,  # False: 기존 index 유지, True: 기존 index 무시
             keys=None, # 계층적 index 사용하려면 keys 튜플 입력
             levels=None,
             names=None, # index의 이름 부여하려면 names 튜플 입력
             verify_integrity=False, # True: index 중복 확인
             copy=True) # 복사
```

```python
import pandas as pd
from pandas import DataFrame
```

```python
df_1 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
'B': ['B0', 'B1', 'B2'],
'C': ['C0', 'C1', 'C2'],
'D': ['D0', 'D1', 'D2']},
index=[0, 1, 2])
df_1
```

```python
df_2 = pd.DataFrame({'A': ['A3', 'A4', 'A5'],
   'B': ['B3', 'B4', 'B5'],
   'C': ['C3', 'C4', 'C5'],
'D': ['D3', 'D4', 'D5']},
   index=[3, 4,5])
df_2
```

```python
df_12_axis0 = pd.concat([df_1, df_2]) # row bind : axis = 0, default
df_12_axis0 
```

```python
df_3 = pd.DataFrame({'E': ['A6', 'A7', 'A8'],
'F': ['B6', 'B7', 'B8'],
'G': ['C6', 'C7', 'C8'],
'H': ['D6', 'D7', 'D8']},
index=[0, 1, 2])
df_3
```

```python
df_13_axis1 = pd.concat([df_1, df_3], axis = 1)
df_13_axis1
```

```python
df_4 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
'B': ['B0', 'B1', 'B2'],
'C': ['C0', 'C1', 'C2'],
'E': ['E0', 'E1', 'E2']},
index=[0, 1, 3])
df_4
```

```python
df_14_inner = pd.concat([df_1, df_4], join='inner') # union, default
df_14_inner
```

```python
df_5 = pd.DataFrame({'A': ['A0', 'A1', 'A2'],
'B': ['B0', 'B1', 'B2'],
'C': ['C0', 'C1', 'C2'],
'D': ['D0', 'D1', 'D2']},
index=['r0', 'r1', 'r2'])
df_5
```

```python
df_6 = pd.DataFrame({'A': ['A3', 'A4', 'A5'],
'B': ['B3', 'B4', 'B5'],
'C': ['C3', 'C4', 'C5'],
'D': ['D3', 'D4', 'D5']},
index=['r3', 'r4', 'r5'])
df_6
```

```python
df_56_with_index = pd.concat([df_5, df_6], ignore_index=False) # default
df_56_with_index 
```

```python
# if you want ignore current index, use 'ignore_index=True'
df_56_ignore_index = pd.concat([df_5, df_6], ignore_index=True)# index 0~(n-1)
df_56_ignore_index
```

```python
# concatenating DataFrames : Construct hierarchical index using 'keys'
df_56_with_keys = pd.concat([df_5, df_6], keys=['df_5', 'df_6'])
df_56_with_keys
```

```python
df_56_with_keys.loc['df_5']
```

```python
# (6) index에 이름 부여하기 : names
df_56_with_name = pd.concat([df_5, df_6],
keys=['df_5', 'df_6'],
names=['df_name', 'row_number'])
df_56_with_name
```
