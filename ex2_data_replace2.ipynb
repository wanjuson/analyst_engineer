```python
import pandas as pd
data1 = pd.read_csv('house_price.csv', encoding = 'utf-8')
data2 = data1[:100]
data2.tail()
```

```python
data2['income_rank'] = data2['income'].rank(method = 'dense')
data2.tail()
```

```python
top10 = data2[data2['income_rank'] == 90]['income']
top10 = float(top10)
```

```python
data2['income2'] = data2['income']
data2.describe()
```

```python
data2['income2'][data2['income2'] > top10] = top10
```

```python
# house_value 496100.000000
fin1 = data2[data2['house_value'] > 496100]['income2'].mean()
fin1
```
