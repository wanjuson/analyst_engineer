```python
## 로지스틱 회귀분석
# 데이터 'breast-cancer-wisconsin.csv'
# 저장 classfication_test.csv
```

```python
# 1.데이터 불러오기
import pandas as pd
data1 = pd.read_csv('breast-cancer-wisconsin.csv', encoding = 'utf-8')
data1.tail()
```

```python
data1.describe()
```

```python
# 2.데이터 전처리
# x,y 나누기
x1 = data1[['Clump_Thickness', 'Cell_Size', 'Cell_Shape', 'Marginal_Adhesion', 'Single_Epithelial_Cell_Size'
          , 'Bare_Nuclei', 'Bland_Chromatin', 'Normal_Nucleoli', 'Mitoses']]
y1 = data1[['Class']]
```

```python
# test, train 데이터 나누기
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, random_state = 1)
```

```python
y1_train.mean()
```

```python
y1_test.mean()
```

```python
# 데이터 정규화 - minmax
from sklearn.preprocessing import MinMaxScaler
minmax_scaler = MinMaxScaler()
minmax_scaler.fit(x1_train)
# train data 
x1_train_minmax_scaler = minmax_scaler.transform(x1_train)
x1_train_minmax_scaler
```

```python
# test data 
x1_test_minmax_scaler = minmax_scaler.transform(x1_test)
x1_test_minmax_scaler
```

```python
# 3.데이터 분석
from sklearn.linear_model  import LogisticRegression
model1 = LogisticRegression()
model1.fit(x1_train_minmax_scaler, y1_train)
```

```python
pred_train = model1.predict(x1_train_minmax_scaler)
pred_train
```

```python
model1.score(x1_train_minmax_scaler, y1_train)
```

```python
pred_test = model1.predict(x1_test_minmax_scaler)
pred_test
```

```python
model1.score(x1_test_minmax_scaler, y1_test)
```

```python
## 예측확률
# train
prob_train = model1.predict_proba(x1_train_minmax_scaler)
prob_train
```

```python
# test
prob_test = model1.predict_proba(x1_test_minmax_scaler)
prob_test
```
