```python
## 5.3 데이터 변환하기
### 가. 복사, 추가, 삭제
#데이터 프레임 전체 복사 (백업)
df_columns = df.copy()
df_columns.tail(2)
```

```python
#데이터 프레임의 열(변수) 이름 확인
df_columns.columns
```

```python
#데이터 프레임 열 필터링
df_columns = df_columns[['size','period','price']]
df_columns.head()
```

```python
### 나. 변수이름 변경, 행추가 및 삭제
# 데이터 열(변수) 이름 수정: period -> time
df_columns.rename(columns={'period' : 'time'}, inplace=True)
df_columns.columns
```

```python
# size를 time으로 나누어 월별 포도크기로 계산하여 growth 변수 생성
df_columns['growth'] = df_columns['size'] / df_columns['time']
df_columns.head()
```

```python
# 특정 변수 삭제: growth 삭제
del df_columns['growth']
df_columns.head(2)
```

```python
### 다. 데이터 케이스 추출
# 특정 케이스 선택1: 집단선택
df_continent_brand = df[(df['continent']==1) & (df['brand']==1)] 
df_continent_brand.tail(2)
```

```python
# 특정 케이스 선택2: 기준값 이상 혹은 이하 선택
df_over_size_period = df[(df['size']>=10) & (df['period']>=30)] 
df_over_size_period.head(2)
```

```python
# 특정 케이스 선택2: 기준값 이상 혹은 이하 선택
df_over_size_period = df[(df['size']>=10)] 
df_over_size_period.count()
```

```python
### 라. 코딩변경
df.head(2)
```

```python
# brand 범주 카운트
df['brand'].value_counts()
```

```python
# - 방법1) replace 사용  안됨

# brand 1, 2, 3 집단을 1, 2로 묶기 (1/2 ->1, 3->2)
recode_brand = {"brand": {1: 1, 2: 1, 3: 2}}

df_recode1 = df.replace(recode_brand)
df_recode1.head()

df_recode1['brand'].value_counts()
```

```python
# - 방법2) 함수정의 사용

# 코딩변경 함수 정의
def brand_groups(series):
    if series==1:
        return 1
    elif series==2:
        return 1
    elif series==3:
        return 2

df['re_brand']=df['brand'].apply(brand_groups)
df.head()
```

```python
# - 방법3) 변경
df2 = df
```

```python
df2['re_brand'] = df['brand'].replace({1: 1, 2: 1, 3: 2})
```
