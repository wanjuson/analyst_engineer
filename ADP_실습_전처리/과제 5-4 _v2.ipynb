```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
iris = pd.read_csv('c:/adp/data/iris.csv')
```

```python
## 산점도 행렬

두 개 이상의 변수가 있는 데이터에서 변수들 간의 산점도를 그린 그래프

여러 변수들 간의 산점도를 한 번에 볼 수 있도록 표현하여 변수들 간의 상관관계를 비교할 수 있음


<seaborn의 pairplot>  
> seaborn.pairplot(data, hue=None, kind='scatter', diag_kind='auto', markers=None, dropna=False, corner=True)
> - data : pandas.DataFrame
> - hue : 데이터의 범주형 변수. 집단을 색으로 구분해서 플롯을 그려줌
> - kind : 'scatter'(산점도), 'kde'(밀도그래프), 'hist'(히스토그램), 'reg'(회귀선)로 구성.
> - diag_kind : 'auto', 'hist', 'kde', None으로 구성
> - markers : 마커 코드
> - corner : True로 할 경우 세모 모양 상관계수 그래프 (디폴트는 False)
```

```python
sns.pairplot(iris, hue = 'target')
```

# Data :  Cars93 데이터 
## X축 : RPM, Y축 Price 산점도 그리기 
## X축 Passengers, Y축 Price, 꺾은선 그래프 그리기 
## Price, RPM 변수 Boxplot 그리기 

```python
Cars = pd.read_csv('c:/adp/data/Cars93.csv')
Cars.info()
```

```python

# X축 : RPM, Y축 Price 산점도 그리기 
# 산점도 그리기 기본
# - plt.scatter(x, y, c, s, alpha)   
#     - x, y : x축과 y축
#     - c: 색
#     - s: 마커의 면적
#     - alpha: 마커색의 투명도
import matplotlib.pyplot as plt
```

```python
plt.scatter(x = Cars['RPM'], y = Cars['Price'], c = 'g', s=100, alpha=0.5)
plt.show()   
```

```python
plt.scatter(x = Cars['RPM'], y = Cars['Price'], c = 'g')
plt.show()   
```

```python
## X축 Passengers, Y축 Price, 꺾은선 그래프 그리기 
plt.plot(Cars['Passengers'], Cars['Price'], linestyle='solid')
plt.show()   
```

```python
Cars_sorted = Cars.sort_values(by = ['Passengers'])
```

```python
plt.plot(Cars_sorted['Passengers'], Cars_sorted['Price'], linestyle='solid')
plt.show()   
```

```python
## Price, RPM 변수 Boxplot 그리기 
```

```python
plt.figure(figsize = (20,5))
plt.subplot(121)
plt.boxplot(Cars['Price'])
plt.subplot(122)
plt.boxplot(Cars['RPM'])
plt.show()
```
