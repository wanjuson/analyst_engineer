```python
# mtcars 데이터셋(mtcars.csv)의 qsec 컬럼을 최소최대 척도(Min-Max Scale)로 변환한 후 0.5보다
# 큰 값을 가지는 레코드 수를 구하시오.
```

```python
# 데이터 불러오기
# 데이터 파일 읽기 예제
import pandas as pd
a = pd.read_csv('mtcars.csv', index_col=0)
a.tail()
```

```python
a2 = a[['qsec']]
a2.tail()
```

```python
# minmax
from sklearn.preprocessing import MinMaxScaler
minmax_scale = MinMaxScaler()
minmax_scale.fit(a2)
a2_scale = minmax_scale.transform(a2)
pd.DataFrame(a2_scale).describe()
```

```python
# count
len(a2_scale[a2_scale > 0.5])
```
