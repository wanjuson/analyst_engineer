```python
## 1.데이터 불러오기
import pandas as pd
data1 = pd.read_csv('house_price.csv', encoding = 'utf-8')
data1.tail()
```

```python
data1.describe()
```

```python
## 2.데이터 전처리
# x와 y변수로 분할
x1 = data1[['housing_age', 'income', 'bedrooms', 'households', 'rooms']]
y1 = data1[['house_value']]
```

```python
x1.tail()
```

```python
y1.tail()
```

```python
# train, test 분할
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1, y1, random_state = 1)
y1_train.mean()
```

```python
y1_test.mean()
```

```python
# 정규화
from sklearn.preprocessing import StandardScaler
standard_scale = StandardScaler()
standard_scale.fit(x1_train)
# train 정규화
x1_train_standard_scale = standard_scale.transform(x1_train)
pd.DataFrame(x1_train_standard_scale).describe()
```

```python
# test 정규화
x1_test_standard_scale = standard_scale.transform(x1_test)
```

```python
## 3.데이터 분석
from sklearn.linear_model import LinearRegression
model1 = LinearRegression()
model1.fit(x1_train_standard_scale, y1_train)
```

```python
x1_train_pred = model1.predict(x1_train_standard_scale)
model1.score(x1_train_standard_scale, y1_train) # R2
```

```python
x1_test_pred = model1.predict(x1_test_standard_scale)
model1.score(x1_test_standard_scale, y1_test) # R2
```

```python
## 5. 결과 저장
y1_train['pred'] = x1_train_pred
y1_test['pred'] = x1_test_pred
```

```python
x1_train_pred = pd.concat([x1_train, y1_train], axis = 1)
x1_test_pred = pd.concat([x1_test, y1_test], axis = 1)
```

```python
x1_pred = pd.concat([x1_train_pred, x1_test_pred], axis = 0)
x1_pred.tail()
```

```python
# 저장
x1_pred.to_csv("regression_test03.csv")
```
