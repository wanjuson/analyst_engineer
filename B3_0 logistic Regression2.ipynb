```python
## 로지스틱 회귀분석
# 데이터 'breast-cancer-wisconsin.csv'
# 저장 classfication_test.csv
```

```python
# 1. 데이터 불러오기
import pandas as pd
data1 = pd.read_csv('breast-cancer-wisconsin.csv', encoding = 'utf-8')
data1.tail()
```

```python
data1.describe()
```

```python
# 2.데이터 전처리
# x,y 나누기
x1 = data1[['code', 'Clump_Thickness', 'Cell_Size', 'Cell_Shape', 'Marginal_Adhesion', 'Single_Epithelial_Cell_Size'
          , 'Bare_Nuclei', 'Bland_Chromatin', 'Normal_Nucleoli', 'Mitoses']]
y1 = data1[['code', 'Class']]
```

```python
print(x1.tail())
print(y1.tail())
```

```python
# train, test 나누기
from sklearn.model_selection import train_test_split
x1_train, x1_test, y1_train, y1_test = train_test_split(x1,y1, random_state = 1)
```

```python
print(y1_train.mean())
print(y1_test.mean())
```

```python
# Z-score 정규화 나누기
from sklearn.preprocessing import StandardScaler
stand_scale = StandardScaler()
stand_scale.fit(x1_train)
# train 정규화
x1_train_stand_scale = stand_scale.transform(x1_train)
pd.DataFrame(x1_train_stand_scale).describe()
```

```python
# test 정규화
x1_test_stand_scale = stand_scale.transform(x1_test)
pd.DataFrame(x1_test_stand_scale).describe()
```

```python
## 3.데이터 분석
# 모델 학습
from sklearn.linear_model import LogisticRegression
model1 = LogisticRegression()
model1.fit(x1_train_stand_scale, y1_train['Class'])
```

```python
pred_train1 = model1.predict(x1_train_stand_scale)
pred_train1
```

```python
model1.score(x1_train_stand_scale, y1_train['Class']) # R2
```

```python
pred_test1 = model1.predict(x1_test_stand_scale)
pred_test1
```

```python
model1.score(x1_test_stand_scale, y1_test['Class']) # R2
```

```python
## 4.분석 결과 저장
y1_train['pred'] = pred_train1
y1_test['pred'] = pred_test1
```

```python
# 에측확률 저장 train
prob_train1 = model1.predict_proba(x1_train_stand_scale)
y1_train['prob'] = prob_train1[:,0]
```

```python
# 에측확률 저장 test
prob_test1 = model1.predict_proba(x1_test_stand_scale)
y1_test['prob'] = prob_test1[:,0]
```

```python
total_train = pd.concat([x1_train, y1_train], axis = 1)
total_test = pd.concat([x1_test, y1_test], axis = 1)
```

```python
total_result = pd.concat([total_train, total_test], axis = 0)
```
