```python
## 2. 제품 200개의 Lot별 불량 제품 수량 데이터에 대해 p관리도를 구하고 시각화하시오.
# n : lot별 생산수량
# p : lot별 불량수량%
# CL(관리중심선) : 전체불량수량/전체생산수량
# 관리상하한선 : p+3*root( p(1-p)/n )
```

```python
df02 = pd.read_csv('C:/adp/data/lot_quality.csv')
df02.info()
```

```python
df02['per'] = df02['불량수량']/ df02['생산수량']
df02['per']
```

```python
## 관리 상한선
df02['up_line'] = df02['per'] + 3*((df02['per']*(1-df02['per'])/df02['불량수량'])**0.5)
df02['under_line'] = df02['per'] - 3*((df02['per']*(1-df02['per'])/df02['불량수량'])**0.5)
```

```python
## 관리 중심선 : 전체 불량 비율
CL = df02['불량수량'].sum()/df02['생산수량'].sum()
CL
```

```python
# (2) 관리도를 시각화하시오.
import matplotlib.pyplot as plt
```
