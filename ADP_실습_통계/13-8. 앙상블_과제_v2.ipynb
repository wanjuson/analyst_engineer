## titanic데이터를 가지고 생존율을 예측하는  아래 3가지의 모델을 만들고 모델을 비교하고 최적 모델의 결과를 해석하세요

### 배깅
### 랜덤포래스트
### 부스팅 



```python
df_t = pd.read_csv("C:/adp/data/titanic.csv")
df_t
```

```python
df_t.info()
```

```python
# 결측치가 있는 컬럼 삭제
isna_col = ['age', 'deck', 'embark_town']
df_t = df_t.drop(isna_col, axis=1)
```

```python
# embarked 결측치가 있는 행 삭제
df_t2 = df_t[df_t['embarked'].isna() == False]
df_t2.info()
```

```python
## 설명력 100%가 나와서 범주형변수 삭제
df_t2 = df_t2.drop(['embarked', 'who', 'alive'], axis=1)
```

```python
# 종속변수, 설명변수 분할
x = df_t2.drop(['survived'], axis=1)
y = df_t2['survived']
```

```python
# 범주형 변수 더미화
# x = pd.get_dummies(x, columns=['sex', 'embarked', 'class', 'who', 'alive'])
x = pd.get_dummies(x, columns=['sex', 'class'])
x.head()
```

```python
# 정규화
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
x_scaled_df = pd.DataFrame(x_scaled, columns = x.columns)
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_scaled_df, y, test_size=0.3, random_state=42, stratify=y) 
```

```python
## 2.1 배깅
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier

param_grid = {'n_estimators':range(110,200,10)}

bag_clf = BaggingClassifier(base_estimator=DecisionTreeClassifier(),
                           max_samples=0.8,
                           max_features=0.9)

grid_bag_clf = GridSearchCV(bag_clf, param_grid, cv=5)
grid_bag_clf.fit(x_train,y_train)
```

```python
grid_bag_clf.best_estimator_
```

```python
from sklearn.metrics import classification_report

y_pred = grid_bag_clf.best_estimator_.predict(x_test)
print('accuracy_score : ', grid_bag_clf.best_estimator_.score(x_test, y_test))

print(f'\n\n{classification_report(y_test, y_pred)}')
```

```python
from sklearn.metrics import plot_roc_curve, roc_auc_score

plot_roc_curve(grid_bag_clf, x_test, y_test)
plt.show()

print("roc_auc_score :", roc_auc_score(y_test, grid_bag_clf.best_estimator_.predict_proba(x_test)[:, 1]))
```

```python
importances = pd.DataFrame(np.mean([i.feature_importances_ for i in grid_bag_clf.best_estimator_.estimators_], axis=0))
feature_importances = pd.concat([pd.DataFrame(x.columns), importances], axis=1)
feature_importances.columns = ['col_names','feature_importances']
feature_importances = feature_importances.sort_values(by='feature_importances', ascending=False).reset_index(drop=True)

plt.barh(feature_importances['col_names'], feature_importances['feature_importances'], align='center')
plt.xlabel('feature_importances', size=10)
plt.ylabel('feature', size=10)
plt.show()
```

```python
# GridSearchCV로 파라미터를 탐색하여 최적의 모델 BaggingClassifier(base_estimator=DecisionTreeClassifier(), max_features=0.9,max_samples=0.8, n_estimators=130)를 찾고 
# 학습한 결과 정확도는 0.805 이고 roc_auc_score는 0.831 의 결과가 나왔다.

# 변수의 중요도를 통해 adult_male, pcclass, sibsp, parch 순으로 survivied에 영향을 주는것으로 파악된다
```
