```python
# 작업형 1유형 예시문제 
# mtcars 데이터셋의 qsec 컬럼을 최소최대 척도(Min Max Scale)로 변환한 후 0.5 보다 큰 값을 가지는 레코드 수를 구하시오.
```

```python
import pandas as pd
mtcar = pd.read_csv("C:/adp/data/mtcars.csv")
mtcar.tail()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(mtcar[['qsec']] )
mtcar['qsec2'] = scaler.transform(mtcar[['qsec']] )
mtcar['qsec2']
```

```python
sum([mtcar['qsec2'] >= 0.5]).sum() # 9개
```
