```python
## 1.데이터
import pandas as pd
# 데이터 불러오기 및 샘플 확인
data = pd.read_csv('house_price.csv', encoding = 'utf-8')
data.head()
```

```python
# 변수별 통계 확인
data.describe()
```

```python
## 2.데이터 전처리
# 데이터 x,y로 분할
x1 = data[['housing_age', 'income', 'bedrooms', 'households', 'rooms']]
print('x1', x1.tail())
y1 = data[['house_value']]
print('y1', y1.tail())
```

```python
# train-test 데이터셋 나누기
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x1, y1, random_state = 1)
print(y_train.mean())
print(y_test.mean())
```

```python
## 정규화 
# train 데이터 정규화
from sklearn.preprocessing import StandardScaler
scaler_standard = StandardScaler()
scaler_standard.fit(x_train)
x_scaler_standard_train = scaler_standard.transform(x_train)
```

```python
# standard 방법으로 정규화한 데이터의 기술통계량 확인
pd.DataFrame(x_scaler_standard_train).describe()
```

```python
# train 데이터 정규화
x_scaler_standard_test = scaler_standard.transform(x_test)
pd.DataFrame(x_scaler_standard_test).describe()
```

```python
## 3. 데이터 분석
## 모델학습
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_scaler_standard_train, y_train)
```

```python
# train 예측
pred_train = model.predict(x_scaler_standard_train)
model.score(x_scaler_standard_train, y_train) # R-square
```

```python
# test 예측
pred_test = model.predict(x_scaler_standard_test)
model.score(x_scaler_standard_test, y_test) # R-square
```

```python
# test RMSE 평가
import numpy as np
from sklearn.metrics import mean_squared_error
MSE = mean_squared_error(y_test, pred_test)
np.sqrt(MSE)
```

```python
## 4. 결과 저장
# train 결과 추가
prob_train = model.predict(x_scaler_standard_train)
y_train[['y_pred']] = prob_train
y_train.tail()
```

```python
# test 결과 추가
prob_test = model.predict(x_scaler_standard_test)
y_test[['y_pred']] = prob_test
y_test.tail()
```

```python
# train 데이터 결합
total_train = pd.concat([x_train, y_train], axis = 1)
total_train.describe()
```

```python
# 테스트 데이터의 특성치(X_test)와 레이블 및 예측치(y_test)를 병함
total_test = pd.concat([x_test, y_test], axis =1)
total_test.describe()
```

```python
# train + test 합치기
total_data = pd.concat([total_train, total_test], axis = 0)
total_data.describe()
```
