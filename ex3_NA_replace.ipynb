```python
# 1) 데이터 전체에서 80% 추출 후 결측치들을 median으로 채우고 표준편차의 변화값:
```

```python
import pandas as pd
data1 = pd.read_csv('house_price.csv', encoding = 'utf-8')
data1.tail()
```

```python
data2 = data1[:100]
data2['income2'] = data2['income']
```

```python
import numpy as np
data2['income2'][data2['income2'] < 3.6] = np.NAN
```

```python
std1 = data2['income2'].std()
std1
```

```python
med1 = data2['income2'].median()
med1
```

```python
data2['income3'] = data2['income2']
data2['income3'][data2['income3'].isna()] = med1
data2
```

```python
std2 = data2['income3'].std()
std2
```

```python
diff = std2 - std1
print(diff)
```
